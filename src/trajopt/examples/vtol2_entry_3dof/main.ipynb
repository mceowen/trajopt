```python
# --- 3D Interactivity (run %matplotlib widget in Jupyter) ---
#%matplotlib widget # ← run this once per notebook session

# --- Retrieve trajectory and parameters ---
z_ref = trajopt_obj.I[-1]["z_ref"]

mission = trajopt_obj.mission
model = trajopt_obj.model
method = trajopt_obj.method

x_init = method.z_init[:, 0]
y_init = method.z_init[:, 1]
z_init = method.z_init[:, 2]

t_init = method.t_init
nu_init = method.nu_init
t_ref  = trajopt_obj.I[-1]["t_ref"]
us_ref  = trajopt_obj.I[-1]["us_ref"]

plt.plot(t_ref * method.nondim['nt'], (z_ref[:, 0] * method.nondim['nd'] - mission.planet['r']) / 1e3)
plt.ylabel('altitude [km]')
plt.xlabel('time [s]')

plt.figure()
plt.plot(t_ref*method.nondim['nt'], (z_ref[:, 3] * method.nondim['nv']))
plt.ylabel('velocity [km/s]')
plt.xlabel('time [s]')

plt.figure()
plt.plot(t_ref*method.nondim['nt'], np.rad2deg(nu_ref[:, 0]))
plt.ylabel('bank angle [deg]')
plt.xlabel('time [s]')

print(f"final time: {t_ref[-1] * method.nondim['nt']}")
```

## 3DoF Entry MSL

```python
from trajopt.trajectory_analyzer import TrajectoryAnalyzer
```

```python
# create trajectory analyzer object
trajopt = TrajectoryAnalyzer("config.yaml")

trajopt.solve()
```

```python
data = trajopt.analyze(analysis_type="standalone", compute_iters=True)
```

```python
trajopt.plot(data, show_iters=True, animate=False)
```

## 3DoF Entry RLV

```python
from trajopt.trajectory_analyzer import TrajectoryAnalyzer
```

```python

config_path = "./config.yaml"

trajopt = TrajectoryAnalyzer(config_path)

trajopt.solve()
```

```python
data = trajopt.analyze(analysis_type="standalone", compute_iters=True)
```

```python
trajopt.plot(data, show_iters=False, animate=False)
```

# Perching Glider

source: https://underactuated.mit.edu/trajopt.html#perching

```python
from trajopt.trajectory_analyzer import TrajectoryAnalyzer
```

```python
# create trajectory analyzer object
traj = TrajectoryAnalyzer("config.yaml")

traj.solve()
```

```python
data = traj.analyze(analysis_type="standalone", compute_iters=True)
```

```python
traj.plot(data, show_iters=True)
```

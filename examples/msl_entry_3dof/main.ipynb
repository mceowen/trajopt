## 3DoF Entry MSL

```python
from trajopt.core.trajectory_analyzer import TrajectoryAnalyzer
```

```python
# create trajectory analyzer object
mission     = "mission.yaml" # "experiments/multiphase_mission.yaml"
model       = "trajopt/models/reentry_3dof.yaml"
method      = "method.yaml"
variations  = "variations.yaml"

trajopt     = TrajectoryAnalyzer(mission, model, method, variations=variations)

trajopt.solve()
```

```python
data = trajopt.analyze(analysis_type="standalone", compute_iters=True)
```

```python
trajopt.plot(data, show_iters=True, animate=False)
```

setup problem and run SCP

```python
# create trajectroy analyzer object
trajopt = TrajectoryAnalyzer("trajopt.yaml")

# run SCP
trajopt.solve()
```

mc analysis

```python
scenario_data = trajopt.analyze()
```

```python
print(scenario_data['autotune']['mc_data'][-1]['iters'][-1]['constraint_data'].keys())
```

```python
import matplotlib.pyplot as plt

plt.figure()

nominal_run = scenario_data['autotune']['mc_data'][0]
solution = nominal_run['iters'][-1]

params = nominal_run['params']

t_opt = solution['t_opt']
z_opt = solution['z_opt']
nu_opt = solution['nu_opt']

t_nl = solution['t_nl']
z_nl = solution['z_nl']
nu_nl = solution['nu_nl']

r_planet = params['mission']['planet']['r']

plt.figure()
plt.plot(t_nl, z_nl[:, 0] - r_planet, 'b-', lw=1, label='NL prop')
plt.plot(t_opt, z_opt[:, 0] - r_planet, 'r.', ms=5, label='z_opt')

# plt.figure()
# plt.plot(t_opt, z_opt[:, 1])
# plt.figure()
# plt.plot(t_opt, z_opt[:, 2])
plt.figure()
plt.plot(t_nl, z_nl[:, 3], 'b-', lw=1, label='NL prop')
plt.plot(t_opt, z_opt[:, 3], 'r.', ms=5, label='z_opt')
# plt.figure()
# plt.plot(t_opt, z_opt[:, 4])
# plt.figure()
# plt.plot(t_opt, z_opt[:, 5])


plt.figure(figsize=(12, 8))
plt.plot(t_opt, nu_opt[:, 0])
plt.plot(t_opt, nu_opt[:, 1])

plt.tight_layout()
plt.show()
```

## 3DoF Entry MSRL

```python
import os

os.environ["TF_CPP_MIN_LOG_LEVEL"] = "3"
from trajopt.core.trajectory_analyzer import TrajectoryAnalyzer
```

```python
mission = "mission.yaml"
model   = "trajopt/models/configs/reentry_3dof.yaml"
method  = "method.yaml"

traj = TrajectoryAnalyzer(mission, model, method)

traj.solve()
```

```python
data = traj.analyze(analysis_type="standalone", compute_iters=True)
```

```python
traj.plot(data, show_iters=True)
```
